# Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to the python environment.

```python
ds=pd.read_csv('cars_data.csv')
```

```python
ds
```

# 2. Check for the null values present in the dataset.

```python
ds.isnull()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(10,10))
sns.set_theme(style='darkgrid')
sns.countplot(x='Buyer Gender',data=ds)
plt.title('Male VS Female Buyers',size=16)
```

From this, we can understand that female buyers have participated more than the male buyers in the sales.

* Female Buyers participated are above 5000 and Male Buyers are below 5000.

# 4. Find the top 5 cars based on their sales price.

```python
ds1=ds.sort_values(by=['Sale Price'],ascending=False).head(5)
```

```python
ds1
```

Full details are available

```python
ds2=ds1[['Make','Model','Sale Price']]
ds2
```

Only the top 5 cars details and their sale price are shown

* Volkswagen Jetta III has the highest sale price.
* 2 models of Lexus also there in top 5
