# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the dataset to the python environment

```python
data=pd.read_csv('StudentsPerformance.csv')
```

# Printing the dataset:

```python
data
```

# 1. Find out how many males and females participated in the test.

```python
count_of_gender=data.groupby('gender').size()
print('Male and female participated in the test are:')
count_of_gender
```

* Plotting a Bar Graph from above data

```python
plt.figure(figsize=(8,8))
sns.set_theme(style='darkgrid')
sns.countplot(x='gender',data=data)
plt.title('Male VS Female Participants',size=20)
```

Findings: 
    
    Number of Female participants is more than the male participants in the test.

# 2. What do you think about the students' parental level of education?

```python
plt.figure(figsize=(20,20))
plt.xlabel("xlabel",fontsize=20)
plt.ylabel("ylabel",fontsize=20)
graph=sns.countplot(x='parental level of education',data=data)
plt.title('Count of Parental level of Education',size=25)
for p in graph.patches:
        graph.annotate('{}'.format(p.get_height()), 
                       (p.get_x()+0.30, 
                        p.get_height()+1))
```

Findings:
    
    * Most parents went to some college and have an Associate's Degree.
    * Least parents had took a Master's Degree.
    * Out of 1000 parents, only 118 have a Bachelor's Degree.
    

# 3. Who scores the most on average for math, reading and writing based on

● Gender

● Test preparation course

Based on Gender

```python
data_gen_mean=data.groupby('gender',as_index=False)[["math score","reading score","writing score"]].mean()
```

```python
print('Mean score based on Gender:\n')
data_gen_mean
```

Based on Test preparation Course

```python
data_course_mean=data.groupby('test preparation course',as_index=False)[["math score","reading score","writing score"]].mean()
```

```python
print('Mean score based on Test Preparation Course:\n')
data_course_mean
```

Findings:

    * Taking the reading and writing scores, female acheived more scores than male.
    * Math score is high for male.

# 4. What do you think about the scoring variation for math, reading and writing based on

● Gender

● Test preparation course

Based on Gender

```python
data_gen_std=data.groupby('gender',as_index=False)[["math score","reading score","writing score"]].std()
```

```python
print('Variation in score values based on Gender\n')
data_gen_std
```

Based on Test preparation Course

```python
data_course_std=data.groupby('test preparation course',as_index=False)[["math score","reading score","writing score"]].std()
```

```python
print('Variation in score values based on Test Preparation Course:\n')
data_course_std
```
